```python
import matplotlib.pyplot as plt
from IPython.display import Audio
import numpy as np
import io
import soundfile as sf
```

```python
import demosongs
import pysynth_beeper
import pysynth
import pysynth_b
import pysynth_c
import pysynth_d
import pysynth_e
import pysynth_p
import pysynth_s
```

```python
ss = demosongs.song4_rh
speed_coeff = 10
song = [(s[0], s[1]*speed_coeff) for s in ss]
```

```python
buffer = io.BytesIO()

pysynth_s.make_wav(song, fn=buffer)

buffer.seek(0)

#wav = buffer.getvalue()
data, samplerate = sf.read(buffer)
```

```python
plt.figure(figsize=(16,4))
plt.plot(data)
```

```python
Audio(data, rate=samplerate)
```
