# post-fork v9 nonces

These are from the blocks that were mined on v9 after the rest of the network upgraded to v10 (at height 1788000)

Data from https://monero.mycryptoapi.com/

Analysis: [Isthmus](https://www.github.com/mitchellpkt)

## Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Import data

```python
v9_nonces = (1770048148,1690395859,549474101,595748996,411047770,214112690,1568836096,54526313,340746207,515924036,1159071430,604006518,977353812,1145046670,964693586,855701013,641920729,843246589,25585649,1266881498,142651466,696299469,742631684,932283410,151104375,1002466126,701215538,339801495,406850212)
v9_miner_upper_bound = 1.35e9
```

## Plot the data

```python
fig = plt.figure(facecolor='w') # transparent background no good on IRC
plt.hist(v9_nonces)
title = 'v9 nonces from 1788000 - ' + str(1788000+len(v9_nonces))
plt.xlabel('nonce value')
plt.ylabel('frequency')
plt.xlim(0,2**32)
plt.title(title)
plt.axvline(x=v9_miner_upper_bound,c='red')
plt.legend(('Top of v9 ASIC(?) range','Nonces mined on v9 after fork'));
plt.text(1e8, 0.1, 'Isthmus');
```
